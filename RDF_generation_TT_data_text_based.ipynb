**RDF creation of tensile test data**

In this notebook, data originating from an arbitrary tensile test is read in as CSV data. Then, this data is transformed to RDF data by creating triples using a text-based approach (simple print statements). Therefore, values tokenized from the CSV are allocated to concepts of the tensile test ontology (TTO) developed in the frame of the PMD project (TTO as PMD application ontology can be found here: https://github.com/materialdigital/application-ontologies/tree/main/tensile_test_ontology_TTO).

```python
# Definition of IRI prefixes (namespaces)
base = "https://w3id.org/pmd/co/"
unit = "http://qudt.org/vocab/unit/"
owl = "http://www.w3.org/2002/07/owl#"

prefix = "https://w3id.org/pmd/ao/tte/"

# Definition of object properties as used for triple creation
PMDinput = "<" + base + "input>" 
PMDoutput = "<" + base + "output>"
PMDcharacteristic = "<" + base + "characteristic>" 
PMDvalue = "<" + base + "value>" 
PMDunit = "<" + base + "unit>"

# Define object properties as owl:ObjectProperties, indeed
print (PMDinput + " a <" + owl + "ObjectProperty> . ")
print (PMDoutput + " a <" + owl + "ObjectProperty> . ")
print (PMDcharacteristic + " a <" + owl + "ObjectProperty> . ")
print (PMDvalue + " a <" + owl + "ObjectProperty> . ")
print (PMDunit + " a <" + owl + "ObjectProperty> . ")

# Creation of triples
for line in CSV.split("\n")[1:]:
    #print(line)
    tok = line.split(";")
    if len(tok)<=1:
        break
    
    # Definition of tokens in accordance with columns of original csv file
    experiment_id = tok[0]    
    testpiece_id = tok[1]
    machine = tok[2]
    temperature = tok[3]
    rp01 = tok[4]
    rp02 = tok[5]
    rm = tok[6]
    operator = tok[13]
    strainrate = tok[14]
    agingtime = tok[25]
    agingtemp = tok[26]
    matstatename = tok[28]

    # Creation of triples for all instances (considering all connections between them). There is about one "block" per instance.
    # Textbased triple creation
    experimentIRI = prefix + experiment_id 
    print ("<" + experimentIRI + "> a <https://w3id.org/pmd/co/ProcessIdentifier> . ")
    print ("<" + experimentIRI + "> " + PMDvalue +  " \"" + experiment_id + "\"  ." )

    processIRI = experimentIRI + "_process" 
    print ("<" + processIRI + "> a <https://w3id.org/pmd/co/TensileTest> . ")
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + experimentIRI + "> ." )

    testpieceID_IRI = prefix + testpiece_id 
    print ("<" + testpieceID_IRI + "> a <https://w3id.org/pmd/co/TestPieceName> . ")
    print ("<" + testpieceID_IRI + "> a <" + base + "Metadata> . ")
    print ("<" + testpieceID_IRI + "> " + PMDvalue +  " \"" + testpiece_id + "\"  ." )
    

    testpieceIRI = testpieceID_IRI  + "_testpiece";
    print ("<" + testpieceIRI + "> a <https://w3id.org/pmd/co/TestPiece> . ")
    print ("<" + testpieceIRI + "> " + PMDcharacteristic +  " <" + testpieceID_IRI + "> ." )

    print ("<" + processIRI + "> " + PMDinput +  " <" + testpieceIRI + "> ." )

    machineIRI = experimentIRI + "_machine"
    print ("<" + machineIRI + "> a <" + base + "TensileTestingMachine> . ")
    print ("<" + machineIRI + "> a <" + base + "Metadata> . ")
    print ("<" + machineIRI + "> " + PMDvalue +  " \"" + machine + "\"^^xsd:string  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + machineIRI + "> ." )

    temperatureIRI = experimentIRI + "_temperature"
    print ("<" + temperatureIRI + "> a <" + base + "EnvironmentalTemperature> . ")
    print ("<" + temperatureIRI + "> a <" + base + "Metadata> . ")
    print ("<" + temperatureIRI + "> a <" + base + "Measurement> . ")
    print ("<" + temperatureIRI + "> " + PMDvalue +  " \"" + temperature + "\"^^xsd:float  ." )
    print ("<" + temperatureIRI + "> " + PMDunit +  " <" + unit + "DEG_C>  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + temperatureIRI + "> ." )

    setTemperatureIRI = experimentIRI + "_setTemperature"
    print ("<" + setTemperatureIRI + "> a <" + base + "EnvironmentalTemperature> . ")
    print ("<" + setTemperatureIRI + "> a <" + base + "Metadata> . ")
    print ("<" + setTemperatureIRI + "> a <" + base + "SetPoint> . ")
    print ("<" + setTemperatureIRI + "> " + PMDvalue +  " \"" + temperature + "\"^^xsd:float  ." )
    print ("<" + setTemperatureIRI + "> " + PMDunit +  " <" + unit + "DEG_C>  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + setTemperatureIRI + "> ." )

    rp01IRI = experimentIRI + "_rp01"
    print ("<" + rp01IRI + "> a <" + base + "Rp01> . ")
    print ("<" + rp01IRI + "> a <" + base + "SecondaryData> . ")
    print ("<" + rp01IRI + "> a <" + base + "Measurement> . ")
    print ("<" + rp01IRI + "> " + PMDvalue +  " \"" + rp01 + "\"^^xsd:float ." )
    print ("<" + rp01IRI + "> " + PMDunit +  " <" + unit + "MegaPa>  ." )
    print ("<" + processIRI + "> " + PMDoutput +  " <" + rp01IRI + "> ." )

    rp02IRI = experimentIRI + "_rp02"
    print ("<" + rp02IRI + "> a <" + base + "Rp02> . ")
    print ("<" + rp02IRI + "> a <" + base + "SecondaryData> . ")
    print ("<" + rp02IRI + "> a <" + base + "Measurement> . ")
    print ("<" + rp02IRI + "> " + PMDvalue +  " \"" + rp02 + "\"^^xsd:float  ." )
    print ("<" + rp02IRI + "> " + PMDunit +  " <" + unit + "MegaPa>  ." )
    print ("<" + processIRI + "> " + PMDoutput +  " <" + rp02IRI + "> ." )

    rmIRI = experimentIRI + "_tensileStrength"
    print ("<" + rmIRI + "> a <" + base + "TensileStrength> . ")
    print ("<" + rmIRI + "> a <" + base + "SecondaryData> . ")
    print ("<" + rmIRI + "> a <" + base + "Measurement> . ")
    print ("<" + rmIRI + "> " + PMDvalue +  " \"" + rm + "\"^^xsd:float  ." )
    print ("<" + rmIRI + "> " + PMDunit +  " <" + unit + "MegaPa>  ." )
    print ("<" + processIRI + "> " + PMDoutput +  " <" + rmIRI + "> ." )

    operatorIRI = experimentIRI + "_operator"
    print ("<" + operatorIRI + "> a <" + base + "Operator> . ")
    print ("<" + operatorIRI + "> a <" + base + "Metadata> . ")
    print ("<" + operatorIRI + "> " + PMDvalue +  " \"" + operator + "\"^^xsd:string  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + operatorIRI + "> ." )

    strainrateIRI = experimentIRI + "_strainRate"
    print ("<" + strainrateIRI + "> a <" + base + "StrainRate> . ")
    print ("<" + strainrateIRI + "> a <" + base + "Metadata> . ")
    print ("<" + strainrateIRI + "> a <" + base + "SetPoint> . ")
    print ("<" + strainrateIRI + "> " + PMDvalue +  " \"" + strainrate + "\"^^xsd:float ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + strainrateIRI + "> ." )

    agingtimeIRI = experimentIRI + "_agingTime"
    print ("<" + agingtimeIRI + "> a <" + base + "AgingTime> . ")
    print ("<" + agingtimeIRI + "> a <" + base + "Metadata> . ")
    print ("<" + agingtimeIRI + "> " + PMDvalue +  " \"" + agingtime + "\"^^xsd:float  ." )
    print ("<" + agingtimeIRI + "> " + PMDunit +  " <" + unit + "HR>  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + agingtimeIRI + "> ." )

    agingtemperatureIRI = experimentIRI + "_agingTemperature"
    print ("<" + agingtemperatureIRI + "> a <" + base + "AgingTemperature> . ")
    print ("<" + agingtemperatureIRI + "> a <" + base + "Metadata> . ")
    print ("<" + agingtemperatureIRI + "> " + PMDvalue +  " \"" + agingtemp + "\"^^xsd:float  ." )
    print ("<" + agingtemperatureIRI + "> " + PMDunit +  " <" + unit + "DEG_C>  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + agingtemperatureIRI + "> ." )

    matstatenameIRI = experimentIRI + "_materialStateName"
    print ("<" + matstatenameIRI + "> a <" + base + "MaterialDesignation> . ")
    print ("<" + matstatenameIRI + "> a <" + base + "Metadata> . ")
    print ("<" + matstatenameIRI + "> " + PMDvalue +  " \"" + matstatename + "\"^^xsd:string  ." )
    print ("<" + processIRI + "> " + PMDcharacteristic +  " <" + matstatenameIRI + "> ." )

    # break

```
